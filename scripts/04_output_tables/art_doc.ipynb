```python
# esta es la [presente ley]
self_references_reglamentos = pd.read_csv('../../data/04_matched/regulation_mentions_matched_self_references.csv')


# Este es su [y su reglamento]
regulations = pd.read_csv('../../data/04_matched/regulation_mentions_matched.csv')

```

```python
# general
regulations_general = pd.read_csv('../../data/04_matched/regulation_mentions_matched_general.csv')
regulations_general.rename(columns={'cdmx_doc_id': 'doc_id'}, inplace=True)
```

```python
# Crear una tabla con las menciones de reglamentos creados por su mencion implicita y que no matchean 
regulations_general_none = regulations_general[regulations_general['doc_id'].isna()]
regulations_general_none
# Guardar tabla
regulations_general_none.to_csv('../../data/05_output/no_match/regulations_general_none.csv', index=False)
```

```python
regulations_general = regulations_general[~regulations_general['art_id'].str.contains('_TRANS', na=False)]

regulations_general = regulations_general[~regulations_general['doc_id'].isna()]
regulations_general['type_of_relation'] = 'mencion_reglamento_general'
regulations_general = regulations_general[['art_id','doc_id','type_of_relation']]

print(regulations_general.head())
print(regulations_general.shape)
```

```python
# codigo
codes = pd.read_csv('../../data/04_matched/code_mentions_matched.csv')
codes.rename(columns={'cdmx_doc_id': 'doc_id'}, inplace=True)
```

```python
# Crear una tabla con las menciones de codigos  implicita y que no matchean 
# Tienne NULL porque el LLM no logró matchear la mención
codes_none = codes[codes['doc_id'].isna()]
codes_none
# Guardar tabla
codes_none.to_csv('../../data/05_output/no_match/codes_general_none.csv', index=False)
```

```python
codes = codes[~codes['art_id'].str.contains('_TRANS', na=False)]
codes = codes[~codes['doc_id'].isna()]
codes['type_of_relation'] = 'mencion_codigo_general'
codes = codes[['art_id','doc_id','type_of_relation']]
codes
```

```python
general_laws = pd.read_csv('../../data/04_matched/law_mentions_matched.csv')
general_laws.rename(columns={'cdmx_doc_id': 'doc_id'}, inplace=True)

general_laws_none = general_laws[general_laws['doc_id'].isna()]
general_laws
# Guardar tabla
general_laws_none.to_csv('../../data/05_output/no_match/general_laws_none.csv', index=False)


general_laws = general_laws[~general_laws['art_id'].str.contains('_TRANS', na=False)]
general_laws = general_laws[~general_laws['doc_id'].isna()]
general_laws['type_of_relation'] = 'mencion_ley_general'
general_laws = general_laws[['art_id','doc_id','type_of_relation']]
general_laws
```

```python
# Codes
codes['total_mentions'] = codes.groupby(['art_id', 'doc_id']).transform('size')
codes = codes.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

# Regulations General (los que se encontraron con un match relajado)
regulations_general['total_mentions'] = regulations_general.groupby(['art_id', 'doc_id']).transform('size')
regulations_general = regulations_general.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

# Regulations 
regulations['total_mentions'] = regulations.groupby(['art_id', 'doc_id']).transform('size')
regulations = regulations.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

# Self References Reglamentos (los que salen implitios (esta ley y sus reglamentos))
self_references_reglamentos['total_mentions'] = self_references_reglamentos.groupby(['art_id', 'doc_id']).transform('size')
self_references_reglamentos = self_references_reglamentos.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

# Self References (la presente ley)
self_references['total_mentions'] = self_references.groupby(['art_id', 'doc_id']).transform('size')
self_references = self_references.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

# Oficial (leyes oficiales)
oficial['total_mentions'] = oficial.groupby(['art_id', 'doc_id']).transform('size')
oficial = oficial.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')


# Leyes general (leyes general)
general_laws['total_mentions'] = general_laws.groupby(['art_id', 'doc_id']).transform('size')
general_laws = general_laws.drop_duplicates(subset=['art_id', 'doc_id'], keep='first')

print("Todas las tablas procesadas")
```

```python
# Unir tablas 
total_tables = pd.concat([
    codes,
    regulations_general, 
    regulations,
    self_references_reglamentos,
    self_references,
    oficial,
    general_laws
], ignore_index=True)
total_tables
```

```python
# Identificar filas duplicadas por (doc_id, art_id)
duplicados = total_tables[total_tables.duplicated(subset=['doc_id', 'art_id', 'type_of_relation'], keep=False)]

print(f"=== FILAS REPETIDAS ===")
print(f"Total de filas repetidas: {len(duplicados)}")
print(f"Pares únicos repetidos: {duplicados.groupby(['doc_id', 'art_id']).ngroups}")

# Ordenar por doc_id y art_id para verlas juntas
duplicados_ordenados = duplicados.sort_values(['doc_id', 'art_id'])
print("\nPrimeras 20 filas repetidas:")
print(duplicados_ordenados[['doc_id', 'art_id', 'type_of_relation']].head(20))
```

```python
total_tables.to_csv('../../data/05_output/art_doc.csv', index=False)
```
