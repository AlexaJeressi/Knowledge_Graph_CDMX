```python
import pandas as pd
```

```python
# Leer la lista de leyes para pegar el nombre de cada ley 
df_leyes = pd.read_csv('../../data/02_catalogs/leyes_hash.csv')
df_leyes = df_leyes[['doc_id','nombre']]
```

```python
df = pd.read_csv('../../data/03_extracted/legal_documents_with_ids.csv')
df.head()
df = df[['art_id','doc_id','text']]
df = df.join(df_leyes.set_index('doc_id'), on='doc_id',how='left')
df.head()

# Crear el catalogo de art_id 
```

```python
# Agregarle las tablas de menciones
df_norm = pd.read_csv('../../data/04_matched/art_norm.csv')
df_materia = pd.read_csv('../../data/04_matched/art_materia.csv')
```

```python
df = df.join(df_norm.set_index('art_id'), on='art_id',how='left')
df = df.join(df_materia.set_index('art_id'), on='art_id',how='left')
df.head()

# Guardar
df.to_csv('../../data/02_catalogs/art_hash.csv', index=False)
```
