```python
import pandas as pd
import os
import sys
import numpy as np
```

```python
df = pd.read_csv('../../data/03_extracted/materia_mentions_extracted.csv')
df = df[~df['art_id'].str.contains('_TRANS', na=False)]
df.head()
```

```python
# Obtener art_ids únicos con mención de ley de materia
art_materia = pd.DataFrame({
    'art_id': df['art_id'].unique(),
    'menciona_ley_materia': True
})
art_materia.head()

# Guardar
art_materia.to_csv('../../data/04_matched/art_materia.csv', index=False)
```
